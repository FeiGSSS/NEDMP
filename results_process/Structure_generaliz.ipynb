```python
import pickle as pkl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def getValue(d):
    vs = [v for k, v in d.items()]
    vs = [[x[0] for x in v] for v in vs]
    return vs
```

```python
with open("../data/synthetic/structure_generalization_gnn.pkl", "rb") as f:
    gnn, dmp = pkl.load(f)

with open("../data/synthetic/structure_generalization_nedmp.pkl", "rb") as f:
    nedmp, dmp = pkl.load(f)

gnn = np.vstack(getValue(gnn))
dmp = np.vstack(getValue(dmp))
nedmp = np.vstack(getValue(nedmp))
```

```python
fig, axs = plt.subplots(1,2,figsize=(8, 4),sharex=False, sharey=True)
cbar_ax = fig.add_axes([.93, .15, .03, .7])

data_names = ["tree", "grid", "barbell", "regular", "er03", "er05", "er08", "complete"]

sns.heatmap(gnn, ax=axs[0], cmap="Greens", vmin= 0, vmax=0.5, linewidth=0.3, cbar=True, 
            cbar_ax=cbar_ax, cbar_kws={'label': 'L1 Error'})
sns.heatmap(nedmp, ax=axs[1], cmap='Greens', vmin= 0, vmax=0.4, linewidth=0.3, cbar=False)

axs[0].xaxis.tick_top()
axs[0].tick_params(left=False, top=False)
axs[0].set_xticklabels(data_names, rotation=45, fontsize=14)
axs[0].set_yticklabels(data_names, rotation=0, fontsize=14)
axs[0].set_title("GNN, L1={:.3f}".format(np.mean(gnn)), y=-0.1, fontsize=15)



axs[1].xaxis.tick_top()
axs[1].set_xticklabels(data_names, rotation=45, fontsize=14)
axs[1].tick_params(left=False, top=False)
axs[1].set_title("NEDMP, L1={:.3f}".format(np.mean(nedmp)), y=-0.1, fontsize=15)
axs[1].tick_params(left=False, bottom=False, right=False)
cbar_ax.figure.axes[-1].yaxis.label.set_size(15)
cbar_ax.tick_params(labelsize=13)

# plt.savefig('generalizarion_across_graph.pdf', bbox_inches = 'tight')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pickle as pkl
```

# 1. Tree

```python
G = nx.balanced_tree(r=3, h=2)
nx.draw_kamada_kawai(G)
with open("../data/synthetic/tree/graph.pkl", "wb") as f:
    pkl.dump(G, f)
print(G.nodes())
```

# 2. Regular graph

```python
G = nx.random_regular_graph(d=3, n=12)
nx.draw_spring(G)
with open("../data/synthetic/regular_graph/graph.pkl", "wb") as f:
    pkl.dump(G, f)
print(G.nodes())
```

# 3. Grid

```python
G = nx.grid_2d_graph(m=3, n=4)
nx.draw_kamada_kawai(G)
node_rename = {n:i for i, n in enumerate(G.nodes())}
G = nx.relabel_nodes(G, node_rename)
with open("../data/synthetic/grid/graph.pkl", "wb") as f:
    pkl.dump(G, f)
print(G.nodes())
```

# 4. Barbell

```python
G = nx.barbell_graph(m1=5, m2=2)
nx.draw_kamada_kawai(G)
node_rename = {n:i for i, n in enumerate(G.nodes())}
G = nx.relabel_nodes(G, node_rename)
with open("../data/synthetic/barbell/graph.pkl", "wb") as f:
    pkl.dump(G, f)
print(G.nodes())
```
